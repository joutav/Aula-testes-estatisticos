```python
import pandas as pd 
```

```python
tmdb = pd.read_csv('dados/tmdb_5000_movies.csv')
tmdb.head()
```

```python
 tmdb.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
fig = plt.figure(figsize = (12,8))


ax = sns.histplot(tmdb.vote_average, kde=True, stat="density")
ax.set_title('Media de votos em finles no TMDB 5000', fontsize = 25)

ax.set_xlabel('Nota média', fontsize = 16)
ax.set_ylabel('Densidade', fontsize = 16)

ax = ax
```

```python
fig = plt.figure(figsize = (12,8))


ax = sns.histplot(tmdb.vote_average)
ax.set_title('Media de votos em finles no TMDB 5000', fontsize = 25)

ax.set_xlabel('Nota média', fontsize = 16)
ax.set_ylabel('Densidade', fontsize = 16)

ax = ax
```

```python
plt.figure(figsize=(12,6))
ax = sns.boxplot(x = tmdb.vote_average, width=0.3)
ax.set_title('Distribuiçao de nota média dos filmes do TMBD 5000', fontsize = 20)
ax.set_xlabel('Nota média do filme', fontsize=14)
```

Filmes com a nota média igual a 0 ou 10 não fazem sentido.

```python
tmdb.query('vote_average == 0').head()
```

```python
tmdb.query('vote_average == 10').head()
```

Filmes que apresentam a nota 10 ou 0 tiveram poucos votos e não é possivel definir uma nota média real a eles

```python
tmdb_com_mais_de_10_votos = tmdb.query('vote_count >= 10')
tmdb_com_mais_de_10_votos.describe()
```

```python
fig = plt.figure(figsize = (12,8))


ax = sns.histplot(tmdb_com_mais_de_10_votos.vote_average, kde=True)
ax.set_title('Media de votos em filmes no TMDB 5000 cujo os filmes apresentam 10 ou mais votos', fontsize = 25)

ax.set_xlabel('Nota média', fontsize = 16)
ax.set_ylabel('Densidade', fontsize = 16)

ax = ax
```

```python
plt.figure(figsize=(12,6))
ax = sns.boxplot(x = tmdb_com_mais_de_10_votos.vote_average, width=0.3)
ax.set_title('Media de votos em filmes no TMDB 5000 cujo os filmes apresentam 10 ou mais votos', fontsize = 25)
ax.set_xlabel('Nota média do filme', fontsize=14)
```

# Análise do Movielens

```python
notas = pd.read_csv('dados/ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby('movieId').mean().rating
nota_media_por_filme.head()
```

```python
fig = plt.figure(figsize = (12,8))


ax = sns.histplot(nota_media_por_filme.values, kde=True, stat='density')
ax.set_title('Media de votos em filmes no movielens 100k', fontsize = 25)

ax.set_xlabel('Nota média', fontsize = 16)
ax.set_ylabel('Densidade', fontsize = 16)

ax = ax
```

```python
quantidade_de_votos_por_filme = notas.groupby('movieId').count()
filmes_com_pelo_menos_10_votos = quantidade_de_votos_por_filme.query('rating >= 10').index
filmes_com_pelo_menos_10_votos.values
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos]
```

```python
fig = plt.figure(figsize = (12,8))


ax = sns.histplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values, kde=True, stat='density')
ax.set_title('Media de votos em filmes no movielens 100k cujo os filmes apresentam 10 ou mais votos', fontsize = 25)

ax.set_xlabel('Nota média', fontsize = 16)
ax.set_ylabel('Densidade', fontsize = 16)

ax = ax
```

```python
plt.figure(figsize=(12,6))
ax = sns.boxplot(x = nota_media_dos_filmes_com_pelo_menos_10_votos.values, width=0.3)
ax.set_title('Media de votos em filmes no movielens 100k cujo os filmes apresentam 10 ou mais votos', fontsize = 25)
ax.set_xlabel('Nota média do filme', fontsize=14)
```

```python
fig = plt.figure(figsize = (12,8))


ax = sns.histplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values, kde=True, stat='density', cumulative=True)
ax.set_title('Media de votos em filmes no movielens 100k cujo os filmes apresentam 10 ou mais votos', fontsize = 25)

ax.set_xlabel('Nota média', fontsize = 16)
ax.set_ylabel('% acomulada de filmes', fontsize = 16)

ax = ax
```

```python
fig = plt.figure(figsize = (12,8))


ax = sns.histplot(tmdb_com_mais_de_10_votos.vote_average, kde=True, stat='density', cumulative=True)
ax.set_title('Media de votos em filmes no TMDB 5000 cujo os filmes apresentam 10 ou mais votos', fontsize = 25)

ax.set_xlabel('Nota média', fontsize = 16)
ax.set_ylabel('% acomulada de filmes', fontsize = 16)

ax = ax
```
